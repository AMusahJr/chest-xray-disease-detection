```python
# Load CSV
```

```python
import pandas as pd

df = pd.read_csv("../data/raw/Data_Entry_2017.csv")
df.head()
```

```python
# Create Binary Label
```

```python
df["Pneumonia"] = df["Finding Labels"].apply(
    lambda x: 1 if "Pneumonia" in x else 0
)

df["Pneumonia"].value_counts()
```

```python
# Patient Level Split
```

```python
from sklearn.model_selection import train_test_split

patients = df["Patient ID"].unique()

train_patients, temp_patients = train_test_split(
    patients, test_size=0.3, random_state=42
)

val_patients, test_patients = train_test_split(
    temp_patients, test_size=0.5, random_state=42
)

train_df = df[df["Patient ID"].isin(train_patients)]
val_df   = df[df["Patient ID"].isin(val_patients)]
test_df  = df[df["Patient ID"].isin(test_patients)]

print(len(train_df), len(val_df), len(test_df))
```

```python
train_df.to_csv("../data/processed/train.csv", index=False)
val_df.to_csv("../data/processed/val.csv", index=False)
test_df.to_csv("../data/processed/test.csv", index=False)

```

```python
import pandas as pd

# Load original splits
train_df = pd.read_csv("../data/processed/train.csv")
val_df = pd.read_csv("../data/processed/val.csv")

# Subsample for faster training
train_df = train_df.sample(n=20000, random_state=42)
val_df = val_df.sample(n=5000, random_state=42)

# Save smaller splits
train_df.to_csv("../data/processed/train_small.csv", index=False)
val_df.to_csv("../data/processed/val_small.csv", index=False)

print("Saved train_small.csv and val_small.csv")
```
